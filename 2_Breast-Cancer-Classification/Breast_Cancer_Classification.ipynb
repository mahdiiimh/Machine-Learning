```python
X = df.drop(columns='Label',axis=1)
Y= df['Label']
```

```python
X
```

```python
Y
```

## Split Data

```python
x_train , x_test , y_train , y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

## Feature scaling

```python
from sklearn.preprocessing import StandardScaler
x_train = StandardScaler().fit_transform(x_train)
```

## Train Model

```python
lr_model = LogisticRegression()
```

```python
lr_model.fit(x_train,y_train)
```

## Evaluation

```python
prediction = lr_model.predict(x_train)
train_acc = accuracy_score(y_train,prediction)
```

```python
print(train_acc)
```

```python
x_test = StandardScaler().fit_transform(x_test)
```

```python
test_prediction = lr_model.predict(x_test)
test_acc = accuracy_score(y_test,test_prediction)
```

```python
print(test_acc)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print("Mean squared error: %.2f" % mean_squared_error(y_test,test_prediction))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test,test_prediction))
```
